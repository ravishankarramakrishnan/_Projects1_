Text Data is High Dimensional, Spasely Distributed and more occurence of words is possible. Text data has Feature Variations that may matter more than others.
Feature Space is ht N Dimensions where our variables are. Feature Extraction is the art of creating a space with the decision boundary

```python
# Convert Data From Input space to Feature Space

# Importing the Libraries and Loading the Corpus

from textDB import load_data
corpus = load_data('hobbies')
```

```python
# Validate Corpus Categories

corpus.categories
```

```python
# Check the Data

corpus.data[:2]
```

```python
# Use Methods from 'corpus' for Data Processing and use '.' to get methods
corpus.target[:2]
```

```python
# Vectorizing the Documents

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
```

```python
# Fit and Transform the Data

docs = vectorizer.fit_transform(corpus.data)
labels = corpus.target
```

```python
# Get the Shape of the data (Sparse Feature Matrix)

docs.shape
```

```python
# Decompose data to 2 or 3 dimensions
# TSNE Helps in Clustering Similar Documents Together (with Distance Measure)

from yellowbrick.text import TSNEVisualizer

tsne = TSNEVisualizer(size=(600,400)) # Height, Width
# Fit the Documents and Labels to Visualiser
tsne.fit(docs, labels)

# Plot the Data
tsne.poof()
```

```python
# Yellowbrics Default uses "Squared Equilidean Distance in their TSNE Model"

# Using Equilidean Distance
# Euclidean Distance is the straight-line distance between 2 points in Euclidean (metric) space


tsne = TSNEVisualizer(size=(600,400), metrics= "euclidean") # Height, Width
# Fit the Documents and Labels to Visualiser
tsne.fit(docs, labels)

# Plot the Data
tsne.poof();
```

### Distance Metrics for Non Numerical Inputs

![Alt-text](https://upload.wikimedia.org/wikipedia/commons/0/08/Manhattan_distance.svg "By User:Psychonaut - Created by User:Psychonaut with XFig, Public Domain, https://commons.wikimedia.org/w/index.php?curid=731390")


**Properties of Metrics**:
1. $\text{d}(a,b) \geq 0$ (**non-negativity**)
2. $\text{d}(a,b)=0 \text{ if and only if } a=b$ (**identity**)
3. $\text{d}(a,b)= \text{d}(b,a)$ (**symmetry**)
4. $\text{d}(a,b) \leq \text{d}(a,c)+ \text{d}(c,b)$ (**traingle inequality**)

**Manhattan Distace**: (aka) CityBlock or Taxicab distance

Consider two vectors $a=(a_1,a_2,...,a_d$ and $b=(b_1,b_2,...,b_d)$ in $\mathbb{R}^d$. The Manhattan distance or $L_1$ distance is $$\text{d}_1(a,b)= \left \|a-b\right \|_1 =  \sum_{i=1}\left | a_i -b_i \right |$$

***Manhattan distance between two points is computed as the sum of the absolute differencces of their Cartesian coordinates.***

The Manhattan distance metric (L1 norm) is consistently preferable than the Euclidean distance metric (L2 norm) for high-dimensional data mining. It may be an ideal choice of metric for sparse text data.

```python
# Using Manhattan Distance

tsne = TSNEVisualizer(size=(600,400), metrics= "cityblock") # Height, Width
# Fit the Documents and Labels to Visualiser
tsne.fit(docs, labels)

# Plot the Data
tsne.poof();
```

### Bray Curtis Dissimilarity and Canberra Distance

Most used in Ecology and Statistical Ecology. Best used in Quantifying the difference between different samples and different size. It Doesnt satisfy Triangle Inequality Axiom

```python
# Bray Curtis Distance

tsne = TSNEVisualizer(size=(600,400), metrics= "braycurtis") # Height, Width
# Fit the Documents and Labels to Visualiser
tsne.fit(docs, labels).poof();
```

```python
# Canberra Distance - Distance between Pairs of points in a Vector Space

tsne = TSNEVisualizer(size=(600,400), metrics= "canberra") # Height, Width
# Fit the Documents and Labels to Visualiser
tsne.fit(docs, labels).poof();
```

### Cosine Distance

It corrects for Any variance in the Length of the Dataset. It is also a very Efficient way to compute distance between Sparse matrix (Throws away the Sparsity/Considers Non Zero Elements).

The angle between two documents:
$$\cos(\overline{X},\overline{Y})=\frac{\sum_{i=1}^{d} x_{i}y_{i}}{\sqrt{\sum_{i=1}^{d}x_{i}^2} \sqrt{\sum_{i=1}^{d}y_{i}^2}}$$

***Cosine similarity measures the similarity between two vectors of an inner product space.***


```python
# Cosine Distance

tsne = TSNEVisualizer(size=(600,400), metrics= "cosine") # Height, Width
# Fit the Documents and Labels to Visualiser
tsne.fit(docs, labels).poof();
```

```python
distance_functions = [
    "braycurtis", "canberra", "chebyshev", "cityblock", "correlation", "cosine",
    "dice", "euclidean", "hamming", "jaccard", "kulsinski", "mahalanobis",
    "matching", "minkowski", "rogerstanimoto", "russellrao", "seuclidean",
    "sokalmichener", "sokalsneath", "sqeuclidean", "yule"
    ]
```

```python
#distance_function = ['sokalmichener', 'rogerstanimoto', 'hamming']

for metric in distance_functions:
    tsne = TSNEVisualizer(size=(600,400), metrics= metric) # Height, Width
    # Fit the Documents and Labels to Visualiser
    tsne.fit(docs, labels).poof();
```

```python
# Omit the Class Labels to Identify if any Patterns Emerge - K Means Clustering

tsne = TSNEVisualizer(labels= ["documents"]) # For Legend
tsne.fit(docs).poof()
```

```python
# Applying K Means Clustering

from sklearn.cluster import KMeans
```

```python
clusters = KMeans(n_clusters= 5)

clusters.fit(docs)

tsne = TSNEVisualizer(size= (800,600))
tsne.fit(docs, ['c{}'.format(c) for c in clusters.labels_]) # We pass Each Cluster Labels as list 
tsne.poof();
```
