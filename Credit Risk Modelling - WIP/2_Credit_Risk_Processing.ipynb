## Pandas Model Estimation

```
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```
# Initiate Logistic Regression

reg = LogisticRegression()
```

```
# Show all columns

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```
# Fit the Training data and Target

reg.fit(inputs_train, loan_data_targets_train)
```

```
# Get the Coefficient

reg.coef_
```

```
# Get the Intercept

reg.intercept_
```

```
# Create Columns and their Coefficients Dataframe

feature_name = inputs_train.columns.values

# Create a Dataframe

summary_table = pd.DataFrame(columns= ["Feature Names"], data = feature_name)

# Create a Value column for Summary Table

summary_table['Coefficients'] = np.transpose(reg.coef_)

# Intercept must be put on top of the List

summary_table.index = summary_table.index + 1

summary_table.loc[0] = ["Intercept", reg.intercept_[0]]

summary_table = summary_table.sort_index()
```
